```python
new_df_dish = df_dish.dropna(subset=['main_ingredient']) # Removing rows that do not have a main ingredient identified
```

```python
new_df_dish.info()
```

```python
scene_df_dish = new_df_dish.drop(['main_ingredient_seafood', 'main_ingredient_steak', 'main_ingredient_chicken'], axis = 1) # dropping extra columns
```

```python
# Fork columns related to different scenes
scene1_df_dish = scene_df_dish.drop(['highest_price', 'lowest_price'], axis = 1)
```

```python
# Turn all date values into numeric, misc columns will turn into NaN
scene1_df_dish['first_appeared']= pd.to_numeric(scene1_df_dish['first_appeared'], errors='coerce')
scene1_df_dish['last_appeared']= pd.to_numeric(scene1_df_dish['last_appeared'], errors='coerce')
```

```python
# Drop all NaN rows
scene1_df_dish = scene1_df_dish.dropna(subset=['first_appeared', 'last_appeared'])
```

```python
# Convert float values to int for the date
scene1_df_dish['first_appeared']= scene1_df_dish['first_appeared'].astype(np.int64)
scene1_df_dish['last_appeared']= scene1_df_dish['last_appeared'].astype(np.int64)
scene1_df_dish.info()
```

```python
pd.set_option("display.max_rows", None)
scene1_df_dish.sort_values(by=['times_appeared'], ascending=False).head(400)
```

```python
scene1_df_dish.sort_values(by=['last_appeared'], ascending=False).head(5)
```

```python
subset_df = scene1_df_dish[scene1_df_dish['times_appeared'] > 5]
column_count = subset_df.count()
print(column_count)
```

```python
scene1_df_dish['main_ingredient'] = scene1_df_dish['main_ingredient'].str.replace('chicken', 'Chicken')
scene1_df_dish['main_ingredient'] = scene1_df_dish['main_ingredient'].str.replace('seafood', 'Seafood')
scene1_df_dish['main_ingredient'] = scene1_df_dish['main_ingredient'].str.replace('steak', 'Steak')
scene1_df_dish['main_ingredient'] = scene1_df_dish['main_ingredient'].str.replace('seafoodsteak', 'Seafood and Steak')
scene1_df_dish['main_ingredient'] = scene1_df_dish['main_ingredient'].str.replace('seafoodchicken', 'Seafood and Chicken')
scene1_df_dish['main_ingredient'] = scene1_df_dish['main_ingredient'].str.replace('steakchicken', 'Steak and Chicken')
scene1_df_dish['main_ingredient'] = scene1_df_dish['main_ingredient'].str.replace('seafoodsteakchicken', 'All')
```

```python
s1_df = scene1_df_dish
```

```python

s1_df['first_appeared'] = pd.to_datetime(s1_df['first_appeared'], format='%Y')
s1_df['last_appeared'] = pd.to_datetime(s1_df['last_appeared'], format='%Y')
s1_df.head(10)
```

```python
s1_df = s1_df.drop(scene1_df_dish[scene1_df_dish['times_appeared'] > 5])
```

```python
# Crazy pivot
dates = s1_df.apply(lambda row: pd.date_range(row['first_appeared'], row['last_appeared']).to_series(), axis=1) \
            .stack() \
            .droplevel(-1)
dates.name = 'Date'

s1_df = s1_df.join(dates)
```
